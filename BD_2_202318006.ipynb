```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.datasets import fetch_20newsgroups
from sklearn.metrics.pairwise import cosine_similarity
```

####You need to implement a TF-IDF vectorizer to convert a collection of documents into TF-IDF vectors. You can use the sklearn’s inbuild dataset fetch_20newsgroups.

```python
df = fetch_20newsgroups(subset='all')
documents = df.data
```

```python
tf_vet = TfidfVectorizer()
tf_mat = tf_vet.fit_transform(documents)
```

#### Create a function to calculate the cosine similarity between two TF-IDF vectors.

```python
def calc_cos_sim(vec1,vec2):
  return cosine_similarity(vec1,vec2)[0][0]
```

```python
def doc_similiarty(input_doc, tf_mat, tf_vet, top_n=5):
  input_vet = tf_vet.transform([input_doc])
  similarities = cosine_similarity(input_vet, tf_mat)
  ranking = similarities.argsort()[0][::-1][:top_n]
  rank_doc  = [documents[i] for i in ranking]
  return rank_doc
```

####Implement a document similarity search function that takes a document as input and returns a list of documents ranked by their similarity to the input document.

```python
a = "DAIICT Gandhinagar Data Science batch"
res = doc_similiarty(a, tf_mat, tf_vet)
```

```python
for i,result in enumerate(res,1):
  print(f"Result {i}:\n{result}\n")
```
