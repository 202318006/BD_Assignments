#### RDD Task -2

```python
text8_rdd = spark.sparkContext.textFile("/FileStore/tables/text8")
```

```python
words_rdd = text8_rdd.flatMap(lambda line: line.lower().split())
words_rdd = words_rdd.map(lambda word: word.strip('.,!?'))

# filtering the words containing the letter 'a'
words_with_a_rdd = words_rdd.filter(lambda word: 'a' in word)

word_count_rdd = words_with_a_rdd.map(lambda word: (word, 1))

# reducing by key to get the frequency of each word
word_frequencies = word_count_rdd.reduceByKey(lambda a, b: a + b)

result = word_frequencies.collect()

for word, frequency in sorted(result):
    print(f"Word '{word}': Frequency {frequency}")
```

### <b>DataFrame Task

```python
 from pyspark.sql.functions import corr, col
```

```python
iris_df = spark.read.json("/FileStore/tables/iris.json")


pearson_correlation = iris_df.select(corr("petalLength", "petalWidth")).first()[0]
print("Pearson Correlation between petalLength and petalWidth:", pearson_correlation)
```

```python
# columns with sepalLength, sepalWidth, and species for rows where petalLength >= 1.4
filtered_df = iris_df.filter(col("petalLength") >= 1.4).select("sepalLength", "sepalWidth", "species")
filtered_df.show()
```
